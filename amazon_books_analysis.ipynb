```python
top_auth=df["Author"].value_counts().head(10)
top_auth
```

```python
auth=sns.barplot(top_auth);
plt.xticks(rotation=90);
```

```python
authrev=df.groupby("Author")["Reviews"].sum().sort_values(ascending=False)
authrev
```

```python
sns.barplot(x=authrev.head(10).index, y=authrev.head(10).values)
plt.xticks(rotation=90);
```
